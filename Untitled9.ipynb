```python
import numpy as np
import pandas as pd
from google.colab import files
data = {
    "item_name": [],
    "Category": [],
    "quantity": [],
    "price_per_unit": [],
    "Date": []
}
item_names = ["Iphone", "Laptop", "Bicycle", "Sound Box", "Tablet", "Smartwatch", "TV", "Headphones", "Mouse", "Keyboard"]
categories = ["smartphone", "computer", "vehicle", "music", "electronics", "wearables", "appliances", "audio", "computer accessories", "computer accessories"]
for i in range(100):
  data["item_name"].append(np.random.choice(item_names))
  data["Category"].append(np.random.choice(categories))
  data["quantity"].append(np.random.randint(1, 100))
  data["price_per_unit"].append(np.random.randint(50, 5000))
  data["Date"].append(f"{np.random.randint(1, 30)}/{np.random.randint(1,13)}/24")

df1 = pd.DataFrame(data)
df1.to_csv("Store.csv", index=False)
files.download('Store.csv')
```

```python
def calculate_total_sales_with_loop(df):
  total_sales_list = []
  for index, row in df.iterrows():
      total_sale = row['quantity'] * row['price_per_unit']
      total_sales_list.append(total_sale)
  df1['total_sale'] = total_sales_list
  return df
df1 = calculate_total_sales_with_loop(df1)

df1
```

```python
summary = {}

summary['total_sales'] = df1['total_sale'].sum()

threshold = 100000
category_sales = df1.groupby('Category')['total_sale'].sum()
high_selling_categories = [category for category, sales in category_sales.items() if sales > threshold]
summary['high_selling_categories'] = high_selling_categories

weekly_sales = df1.groupby('Week')['total_sale'].sum().to_dict()
summary['weekly_sales'] = weekly_sales

summary
```

```python

def display_summary(summary):
    """Displays the sales summary in a readable format."""

    print("Sales Summary:")
    print(f"Total Sales: {summary['total_sales']}")
    print("\nHigh-Selling Categories:")
    if summary['high_selling_categories']:
        for category in summary['high_selling_categories']:
            print(f"- {category}")
    else:
        print("No categories exceeded the sales threshold.")

    print("\nWeekly Sales:")
    for week, sales in summary['weekly_sales'].items():
        print(f"Week {week}: {sales}")
display_summary(summary)
```
