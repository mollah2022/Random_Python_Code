```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# NUmpy,pamdas,python,sklearn,(matplotlib,seaborn == same ),(keras,tensorflow,pytorch == same )
```

```python
def unitStep(v):
    if v >= 0:
        return 1
    else:
        return 0
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def perceptronModel(x, w, b, activation='unitStep'):
    v = np.dot(w, x) + b
    if activation == 'unitStep':
        y = unitStep(v)
    elif activation == 'sigmoid':
        y = sigmoid(v)
    return y
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 100
```

```python
# Training the perceptron with unit step activation
w_unit, b_unit = np.zeros(2), 0
for _ in range(epochs):
    for i in range(X.shape[0]):
        y_pred = perceptronModel(X[i], w_unit, b_unit)
        error = y[i] - y_pred
        w_unit += learning_rate * error * X[i]
        b_unit += learning_rate * error
```

```python
# Training the perceptron with sigmoid activation
w_sig, b_sig = np.zeros(2), 0
for _ in range(epochs):
    for i in range(X.shape[0]):
        y_pred = perceptronModel(X[i], w_sig, b_sig, activation='sigmoid')
        error = y[i] - y_pred
        w_sig += learning_rate * error * X[i]
        b_sig += learning_rate * error
```

```python
# Plotting the decision boundaries
plt.figure(figsize=(10, 5))
```

```python
# Subplot 1: Unit Step Function
plt.subplot(1, 2, 1)
x_values = np.linspace(0, 1, 100)
y_values = - (w_unit[0] * x_values + b_unit) / w_unit[1]
plt.plot(x_values, y_values, label='Decision Boundary')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Perceptron with Unit Step Function")
plt.xlabel("x1")
plt.ylabel("x2")
```

```python
# Subplot 2: Sigmoid Function
plt.subplot(1, 2, 2)
x_values = np.linspace(0, 1, 100)
y_values = - (w_sig[0] * x_values + b_sig) / w_sig[1]
plt.plot(x_values, y_values, label='Decision Boundary')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Perceptron with Sigmoid Function")
plt.xlabel("x1")
plt.ylabel("x2")
```

```python
plt.tight_layout()
plt.show()
```
